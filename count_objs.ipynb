```python
import mahotas as mh
```

```python
drinks = mh.imread("COUNT_THIS.jpg")
# drinks = drinks.astype(np.uint8)
drinks = drinks[:,:]
```

```python
pylab.imshow(drinks)
```

```python
%matplotlib inline
```

```python
drinks.shape
```

```python
ts = mh.thresholding.rc(drinks)
```

```python
pylab.imshow(drinks> ts)
```

```python
drinks_filtered = mh.gaussian_filter(drinks,3)
```

```python
pylab.imshow(drinks_filtered )
```

```python
labeled,nr_objects = mh.label(drinks_filtered > ts)
```
